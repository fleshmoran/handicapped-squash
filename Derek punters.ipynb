```python
import pandas as pd
import numpy as np
import datetime
from datetime import date
import random
import matplotlib.pyplot as plt
import time
import sys
```

```python
class MATCHSIM:
    def __init__(self, firstToGames=None, winScore=None, winDiff=None, loseDiff=None):

        #Variable Setting
        self.firstToGames=firstToGames
        self.winScore=winScore
        self.winDiff=winDiff
        self.loseDiff=loseDiff

    def oneGame(self, myStart, oppStart):
        myScore = myStart
        oppScore = oppStart
        
        while max(myScore, oppScore) < self.winScore:
            if np.random.rand() <= self.myPointWinProb:
                #we win a point
                myScore += 1
            else:
                oppScore += 1
            
        #print (myScore,oppScore)
        
        if myScore == self.winScore:
            return 1
        else:
            return 0
        
    def oneMatch(self):
        myStart = self.myStart
        oppStart = self.oppStart
        myGames = 0
        oppGames = 0
        
        while max(myGames, oppGames) < self.firstToGames:
            if self.oneGame(myStart, oppStart) == 1:
                myGames += 1
                myStart += self.winDiff
                oppStart += self.loseDiff
            else:
                oppGames += 1
                myStart += self.loseDiff
                oppStart += self.winDiff
                
        if myGames == self.firstToGames:
            #print ("We won")
            return 1
        else:
            #print ("We lost :(")
            return 0
            
    def play(self, myExpFlat=None, oppExpFlat=None, myStart=None, oppStart=None, runs=range(0,10) ):    
        myWins = 0
        counter = 0
        self.myPointWinProb = myExpFlat / (myExpFlat + oppExpFlat) 
        
        for i in runs:
            self.myStart = myStart
            self.oppStart = oppStart
            myWins += self.oneMatch()
            counter += 1
            
        return round( myWins / counter * 100,1);
        
```

```python
match=MATCHSIM(firstToGames=3, winScore=15, winDiff=-1, loseDiff=2)
runtot = 500
j_low = -5
j_ups = 10

low_rng = 8

for mystrt in range (-5,6):
    frame = []

    for j in range(j_low,j_ups+1):
        lst = []
        for i in range(low_rng,16):
            result = match.play( myExpFlat=15, oppExpFlat=i, myStart=mystrt, oppStart=j, runs=range(0,runtot))
            #print(result)
            lst.append(result)
        #print(lst)
        frame.append(lst)
        #print (frame)
    
    df = pd.DataFrame(frame, columns = range(low_rng,16), index = range(j_low,j_ups+1) )    
    #df = pd.DataFrame(frame)    
    #print(df)


    from IPython.display import display
    display(df.head(j_ups))
    print ("Above is my Start = ", mystrt)

    
```
